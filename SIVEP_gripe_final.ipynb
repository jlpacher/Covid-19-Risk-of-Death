O objetivo desse trabalho é fornecer informações ao nível individual sobre a probabilidade de óbito na hipótese de se contaminar com o vírus SARS2 causador da COVID19.   

Com a decretação da pandemia essa informação poderia ser usada para o indivíduo decidir se expor ao contato com outros ou se resguardar em quarentena.   

Também poderia ser usada pelo governo para adoção de estratégia de isolamento vertical que implica manter as pessoas dos grupos de riscos, como idosos, diabéticos e hipertensos, isolados. Em oposição ao isolamento horizontal que significa manter o maior número possível de pessoas em suas residências e, portanto, isoladas, mas com efeito negativo considerável na economia especialmente se longo.


## DESCRIÇÃO DO DATASET: 

Estudo da mortalidade do COVID-19 baseado no dataset público SIVEP-Gripe (Sistema de Informação de Vigilância Epidemiológica da Gripe) contendo dados epidemiológicos de 291.130 pacientes de diferentes estados do Brasil.

O estudo considerou os pacientes hospitalizados e que testaram positivo para o RT-PCR SARS coronavirus 2.

Foram diagnosticados e estudados 97.924 casos de COVID-19 no intervalo de 1 de Março a 16 de Junho de 2020.

Cada entrada tem 134 atributos, incluindo idade, sexo, etnia e comorbidades entre outros.

Muitos atributos foram desprezados no estudo por não estarem alinhados ao objetivo do estudo ou por não estarem disponíveis no momento da realização da previsão("leakage"), como por exemplo os sintomas dos pacientes que já haviam adquirido a doença.

### ENSEMBLE VOTING

Vamos fazer um Ensemble Voting na tentativa de melhorar os resultados individuais de cada modelo usado.

```python
%%time
# modelos individuais otimizados a serem executados
clf1 = LogisticRegression(fit_intercept = True, penalty = best_penalty, C = best_C,
                            class_weight = best_class_weight, max_iter=5000
                         )

clf2 = GradientBoostingClassifier(random_state=42, n_estimators=best_n_estimators, max_depth=best_max_depth,
                                  min_samples_split=best_min_samples_split, min_samples_leaf=best_min_samples_leaf,
                                  max_features=best_max_features, subsample=best_subsample, learning_rate=learning_rate
                                 )                          
                          
clf3 = CatBoostClassifier(iterations=10000, early_stopping_rounds = 10, cat_features=cat_features, loss_function='Logloss',
                          custom_loss=['Recall', 'Accuracy'], depth=best_depth, l2_leaf_reg=best_l2_leaf_reg,
                          random_strength=best_random_strength, bagging_temperature=best_bagging_temperature,
                          boosting_type=best_boosting_type, rsm=best_rsm, one_hot_max_size=best_one_hot_max_size,
                          leaf_estimation_iterations=best_leaf_estimation_iterations,
                          max_ctr_complexity=best_max_ctr_complexity, subsample=best_subsample, verbose=False
                          )                          

for clf in (clf1, clf2, clf3):
    clf.fit(X_train, y_train)

print('*****************ClassifiersDone*****************')
```

```python
# Resultados dos Modelos
models=['Logistic', 'GradientBoost', 'CatBoost']
Trainning_Accuracy=[clf1.score(X_train,y_train),
                    clf2.score(X_train, y_train),
                    clf3.score(X_train, y_train),
                    ]
Validation_Accuracy=[clf1.score(X_test, y_test),
                    clf2.score(X_test, y_test),
                    clf3.score(X_test, y_test),
                    ]

pd.DataFrame(list(zip(Trainning_Accuracy, Validation_Accuracy)), 
                  columns =['Trainning_Accuracy', 'Validation_Accuracy'],
                  index=models)
```

```python
print('5-fold cross validation:\n')
# labels para facilitar a impressão
models=['Logistic', 'GradientBoost', 'CatBoost']

# for que irá executar os modelos definidos com k-fold cross validation de 5
for clf, label in zip([clf1, clf2, clf3], models):
    scores = model_selection.cross_val_score(clf, X_train, y_train, 
                                              cv=5, 
                                              scoring='accuracy')
    print("Accuracy: %0.2f (+/- %0.2f) [%s]"
          % (scores.mean(), scores.std(), label))
```

```python
# instanciação do Ensemble Voting Hard
models=['Logistic', 'GradientBoost', 'CatBoost']
voting_clf_hard = VotingClassifier(estimators = [(models[0], clf1),
                                                 (models[1], clf2),
                                                 (models[2], clf3)
                                                ],
                                   voting = 'hard')
```

```python
# instanciação do Ensemble Voting Soft
voting_clf_soft = VotingClassifier(estimators = [(models[0], clf1),
                                                 (models[1], clf2),
                                                 (models[2], clf3)
                                                ],
                                   voting = 'soft')
```

```python
# execução dos Ensembles
labels_new = ['Voting_Classifier_Hard', 'Voting_Classifier_Soft']

for (clf, label) in zip([voting_clf_hard, voting_clf_soft], labels_new):
    scores = model_selection.cross_val_score(clf, X_test, y_test, cv=5,
            scoring='accuracy')
    scores2 = model_selection.cross_val_score(clf, X_test, y_test, cv=5,
            scoring='recall')
    
    print("Accuracy: %0.2f (+/- %0.2f) [%s]"
          % (scores.mean(), scores.std(), label))
   
    print("Recall: %0.2f (+/- %0.2f) [%s]"
          % (scores2.mean(), scores2.std(), label))
    
```
